```python
import pandas as pd
import glob
from sqlalchemy import create_engine
# PyMySQL
import pymysql
pymysql.install_as_MySQLdb()
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
engine = create_engine('mysql://root:June-17-94@localhost:3306/stock_db')
```

```python
stock = 'AAPL'
```

```python
myQuery = f'select * from stock_db.master WHERE Name="{stock}"'
df = pd.read_sql_query(myQuery, engine)
df.head()
```

```python
df2 = df.loc[df['index']%100==0, :]
```

```python
df2.head(20)
```

```python
date_list = df2['Date'].tolist()
close_list = df2['Close'].tolist()
```

```python
plt.figure(figsize=(20,5))
plt.title("Stock price over time")
plt.ylabel("Stock Price")
plt.xlabel("Date")
plt.xticks(rotation=90)
plt.plot(date_list, close_list)
plt.yticks()
plt.gcf().subplots_adjust(bottom=0.3)
#plt.autoscale()
plt.show()
plt.clf()      
```

```python
df2['Volume'] = df2['Volume'].astype(int)
```

```python
vol_date_list = df2['Date'].tolist()
vol_close_list = df2['Volume'].tolist()
```

```python
plt.figure(figsize=(20,5))
plt.title("Volume over time")
plt.ylabel("Volume in shares")
plt.xlabel("Date")
plt.xticks(rotation=90)
plt.ticklabel_format(style='plain')
plt.bar(vol_date_list, vol_close_list )
plt.subplots_adjust(bottom=0.3)
plt.show()
```
